## Aplikasi sewa kamar kos
- Melihat kamar kos yang tersedia
- Menyewa kamar
- Melihat pemesanan kamar
- Batal menyewa

````python
# Definisi konstan
AWAL_KAMAR=[104,103,102,101]
AWAL_JENIS =["Panorama","Studio","Double", "Single"]
AWAL_HARGA=[2000000,1000000,400000,200000]
AWAL_STATUS=["Tersedia","Tersedia","Tersedia","Tersedia"]

# Definisi variabel
list_kamar = AWAL_KAMAR.copy()
list_jenis = AWAL_JENIS.copy()
list_harga = AWAL_HARGA.copy()
list_status = AWAL_STATUS.copy()
list_sewa : list[str] = ["", "", "", ""]

print(list_kamar, list_jenis, list_harga, list_status)
````

````python
# Definisi fungsi
def hitung_harga(kamar:int, lama:int) -> int:
    if kamar in list_kamar: # jika kamar tidak ada dalam list
        index = list_kamar.index(kamar)
        return list_harga[index] * lama
    else:
        print("Kamar tidak ditemukan")
        return -1 # -1 menandakan error

def cek_status(kamar:int) -> str:
    if kamar in list_kamar: # jika kamar tidak ada dalam list
        index = list_kamar.index(kamar)
        return list_status[index]
    else:
        print("Kamar tidak ditemukan")
        return "Kamar tidak ditemukan" # -1 menandakan error

def sewakamar(kamar:int, nama:str) -> bool:
    if kamar in list_kamar: # jika kamar tidak ada dalam list
        index = list_kamar.index(kamar)
        if list_status[index] == "Tersedia":
            list_status[index] = "Disewa"
            list_sewa[index] = nama
            return True
        else:
            print("Kamar sudah disewa")
            return False
    else:
        print("Kamar tidak ditemukan")
        return False

def batalsewa(kamar:int) -> bool:
    if kamar in list_kamar: # jika kamar tidak ada dalam list
        index = list_kamar.index(kamar)
        if list_status[index] == "Disewa":
            list_status[index] = "Tersedia"
            list_sewa[index] = ""
            return True
        else:
            print("Kamar sudah tersedia")
            return False
    else:
        print("Kamar tidak ditemukan")
        return False
    
def lihatkamar():
    print("%-10s%-10s%-10s%-10s" % ("Kamar", "Jenis", "Harga", "Status"))
    for i in range(len(list_kamar)):
        print("%--10s%-10s%-10s%-10s" % (list_kamar[i], list_jenis[i], list_harga[i], list_status[i]))

def lihatsewa():
    print("%-10s%-10s%-10s%-10s" % ("Kamar", "Jenis", "Harga", "Status"))
    for i in range(len(list_kamar)):
        if list_status[i] == "Disewa":
            print("%--10s%-10s%-10s%-10s" % (list_kamar[i], list_jenis[i], list_harga[i], list_status[i]))
        
def menu_input():
    while True:
        print("""
        Aplikasi sewa kamar Apartemen Bekom
        1. Kamar kos yang tersedia 
        2. Menyewa Kamar kos  
        3. Melihat pemesanan kamar kos 
        4. Batal menyewa kamar 
        5. Keluar 
        """)

        input_user = int(input("Masukkan pilihan saudara: "))

        # Cek apakah input_user adalah angka 1-5
        if input_user < 1 or input_user > 5:
            print("Pilihan tidak valid, silakan coba lagi.")
            continue
        else:
            return input_user
````

````python
# Main program
def run():
    while True:
        pilihan = menu_input()
        
        if pilihan == 1:
            lihatkamar()
        elif pilihan == 2:
            kamar_input = int(input("Masukkan pilihan kamar saudara: "))
            if kamar_input not in list_kamar:
                print("Kamar tidak ditemukan")
                continue
            elif cek_status(kamar_input) == "Disewa":
                print("Kamar sudah disewa")
                continue
            
            nama_input = input("Masukkan nama: ")

            lama_input = int(input("Masukkan jumlah hari penyewaan: "))

            if sewakamar(kamar_input, nama_input):
                total_harga = hitung_harga(kamar_input, lama_input)
                print(f"Kamar {kamar_input} berhasil disewa oleh {nama_input} dengan harga {total_harga}")
        elif pilihan == 3:
            lihatsewa()
        elif pilihan == 4:
            kamar_input = int(input("Masukkan nomor kamar yang ingin dibatalkan: "))
            if batalsewa(kamar_input):
                print(f"Kamar {kamar_input} telah dibatalkan.")
        elif pilihan == 5:
            print("Terima kasih telah menggunakan App sewa kamar, selamat dan sehat")
            break
        else:
            print("Program rusak, seharusnya tidak bisa sampai sini")
````

````python
run()
````
