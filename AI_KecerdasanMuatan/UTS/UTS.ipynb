```python

import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
# Read cuci_tanga.csv
cuci_tangan_df = pd.read_csv("cuci_tangan.csv")
# Remove line 1
cuci_tangan_df = cuci_tangan_df.drop([0])
# Rename columns
cuci_tangan_df.columns = ['Wilayah', 'Persentase']
# Remove all rows with missing values ('-') or NAN
cuci_tangan_df = cuci_tangan_df[cuci_tangan_df.Persentase != '-']
cuci_tangan_df = cuci_tangan_df.dropna()
# drop last row
cuci_tangan_df = cuci_tangan_df.drop(cuci_tangan_df.index[-1])
# Convert 'Persentase' to float
cuci_tangan_df['Persentase'] = cuci_tangan_df['Persentase'].astype(float)

print(cuci_tangan_df.head())
cuci_tangan_df.dtypes
cuci_tangan_df.describe()
```

```python
# Read pengeluaran_daging.csv
cuci_tangan_df = pd.read_csv("pengeluaran_daging.csv")
# Remove line 1
cuci_tangan_df = cuci_tangan_df.drop([0])
# only keep the column 0 and the average of column 8
cuci_tangan_df = cuci_tangan_df.iloc[:, [0, 8, 9, 10, 11, 12]]
# Rename columns
cuci_tangan_df.columns = ['Wilayah', 'Harga1', 'Harga2', 'Harga3', 'Harga4', 'Harga5']
# Remove all rows with missing values ('-') or NAN
cuci_tangan_df = cuci_tangan_df[cuci_tangan_df.Harga1 != '-']
cuci_tangan_df = cuci_tangan_df[cuci_tangan_df.Harga2 != '-']
cuci_tangan_df = cuci_tangan_df[cuci_tangan_df.Harga3 != '-']
cuci_tangan_df = cuci_tangan_df[cuci_tangan_df.Harga4 != '-']
cuci_tangan_df = cuci_tangan_df[cuci_tangan_df.Harga5 != '-']
cuci_tangan_df = cuci_tangan_df.dropna()
# drop last row
cuci_tangan_df = cuci_tangan_df.drop(cuci_tangan_df.index[-1])
# Convert 'Persentase' to float
cuci_tangan_df['Harga1'] = cuci_tangan_df['Harga1'].astype(float)
cuci_tangan_df['Harga2'] = cuci_tangan_df['Harga2'].astype(float)
cuci_tangan_df['Harga3'] = cuci_tangan_df['Harga3'].astype(float)
cuci_tangan_df['Harga4'] = cuci_tangan_df['Harga4'].astype(float)
cuci_tangan_df['Harga5'] = cuci_tangan_df['Harga5'].astype(float)
# Calculate the average of columns 1-5
cuci_tangan_df['Harga'] = cuci_tangan_df.iloc[:, 1:6].mean(axis=1)
# Drop columns 1-5
cuci_tangan_df = cuci_tangan_df.drop(cuci_tangan_df.columns[1:6], axis=1)

print(cuci_tangan_df.head())
cuci_tangan_df.dtypes
cuci_tangan_df.describe()
```
